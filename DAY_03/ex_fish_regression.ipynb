### 목표 : 무게에 따른 길이를 예측해주는 모델

-   데이터 : fish.csv
-   피쳐/특성 : 무게
-   라벨/타겟 : 길이


(1) 모듈 로딩 및 데이터 준비


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# 데이터 파일
data_file = "../data/fish.csv"
```

```python
# 데이터 준비
fishDF = pd.read_csv(data_file, usecols=[0, 1, 2])
fishDF.info()
```

```python
# 학습에 사용할 데이터 추출 => Perch 행만 추출
perchDF = fishDF[fishDF.Species == "Perch"]

# 인덱스 정리
perchDF.reset_index(drop=True, inplace=True)
```

```python
# 추출된 데이터 기본 정보 확인
perchDF.info()
```

```python
perchDF.head(3)
```

```python
# 수치 컬럼의 데이터 정보 확인
perchDF.describe()
```

(3) 데이터 전처리 <hr>

-   결측치, 이상치, 중복값 처리
-   데이터 분포, 컬럼 분포, 최빈값, 고유값


(3-1) 데이터 분포


```python
# 무게와 길이에 대한 시각화
plt.plot(perchDF.Weight, perchDF.Length, "go")
plt.xlabel("Weight")
plt.ylabel("Length")
plt.title("Weight & Length")
plt.show()
```

```python
# 무게와 길이에 대한 상관계수
perchDF.corr(numeric_only=True)
```

```python
## [분석 결과] 무게에 따른 길이는 비례관계
# 무게 + 연산 ===> 길이
perchDF.iloc[:3]
```

```python
model = (perchDF.Weight * 1 / 5) + 7
model[:3]
```

```python
(model - perchDF.Length).abs().sum()
```

```python
model = (perchDF.Weight * 1 / 5) + 7.2
model[:3]
```

(4) 학습 진행 <hr>

-   학습 방법 : 지도학습 + 회귀(예측) => 선형회귀 LinearRegression


```python
# 모델 인스턴스 생성
model = LinearRegression()
```

```python
# 학습
model.fit(perchDF[["Weight"]], perchDF["Length"])
```

```python
# 학습 후 속성 확인
print(f"피쳐에 할당되는 가중치 계수 : {model.coef_}")
print(f"수식에 사용되는 절편 계수 : {model.intercept_}")
```

```python
(perchDF.Weight * model.coef_) + model.intercept_
```

(5) 테스트 <hr>


```python
model.score(perchDF[["Weight"]], perchDF["Length"])
```

```python
y_pre = model.predict(perchDF[["Weight"]])
y_pre
```

```python
y_pre - perchDF.Length
```
